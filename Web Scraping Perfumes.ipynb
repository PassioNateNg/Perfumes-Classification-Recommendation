```python
from bs4 import BeautifulSoup as BS
import requests
import pandas as pd
from selenium import webdriver
import requests
import shutil
import urllib.request
from PIL import Image
#from IPython.display import Image
```

```python
driver = webdriver.Chrome("/Users/gigichoi/Downloads/chromedriver 2")
url = 'https://www.sephora.com/beauty/best-selling-perfume'
driver.get(url)

sp = BS(driver.page_source, 'html.parser')

brand = sp.find_all('span','css-182j26q')
product_name = sp.find_all('span','css-pelz90')
price = sp.find_all('span','css-0')
review_count = sp.find_all('span','css-1dk1ux')
rating = sp.find_all('div','css-jp4jy6')
image = sp.find_all('img', 'css-1rovmyu eanm77i0')
product_link = sp.find_all('a','css-ix8km1')

brandloop = [brands.text for brands in brand]
product_nameloop = [products.text for products in product_name]
priceloop = [prices.text for prices in price]
review_countloop = [reviews.text for reviews in review_count]
ratingloop = [re['aria-label'] for re in rating]
imageloop = ["www.sephora.com" + img['src'] for img in image]
product_linkloop = ["www.sephora.com" + products['href'] for products in product_link]


priceloop = priceloop [1:]



print(len(brandloop))
print(len(product_nameloop))
print(len(priceloop))
print(len(review_countloop))
print(len(ratingloop))
print(len(imageloop))
print(len(product_linkloop))
```

```python
data = {
    'Brand': brandloop,
    'Product name': product_nameloop,
    'Price': priceloop,
    'Review count':review_countloop,
    'Rating':ratingloop,
    'Image':imageloop,
    'Product link':product_linkloop
}
```

```python
df = pd.DataFrame(data)
```

```python
df = df.replace('\n','', regex=True)
```
